```python
#Question No.5
class Circular_Queue():
    def __init__(self,size): #a constructor
        self.queue=[]
        self.front=-1
        self.rear=-1
        self.size = size
        
    def isfull(self):
        if(self.front==0 and self.rear== self.size-1):
            return True
        elif(self.front==self.size):
            return True
        return False
    
    def isEmpty(self):
        if(self.front==-1 and self.rear==-1):
            return True
        return False
    
    def enqueue(self,val):
        if(self.isfull()):
            print("The Queue is Full")
        else:
            if(self.front == -1):
                self.front = self.front + 1
            self.rear = (self.rear+1)%self.size
            self.queue.append(val)
            print("Value inserted ",val)
   
    def dequeue(self):
        if(self.isEmpty()):
            print("The Queue is Empty")
        else:
            val = self.queue[self.front]
            if(self.front == self.rear):
                self.front = -1
                self.rear = -1
            else:
                self.front = (self.front + 1) % self.size
                
            print("Value deleted ",val)
    
    def display(self):
        if(self.isEmpty()):
            print("The Queue is Empty")
        else:
            for i in range(self.front,self.rear+1):
                print(self.queue[i])

    def Maximum(self):
        if(self.isEmpty()):
            print("The Queue is Empty")
        else:
            max = self.queue[self.front]
            for i in range(self.front,self.rear+1):
                if(max<self.queue[i]):
                    max = self.queue[i]
        print("The maximum value in the Queue is : ",max)
        
c = Circular_Queue(5)
c.enqueue(2)
c.enqueue(3)
c.enqueue(5)
c.enqueue(20)
c.enqueue(38)
c.enqueue(55)
c.enqueue(22)
c.enqueue(31)
c.enqueue(52)

c.display()
c.dequeue()
c.display()

c.Maximum()
```

```python
#Question No.6
from queue import Queue
class Stack_from_Queue():
    def __init__(self):
        self.Q1 = Queue()
        self.Q2 = Queue()
        self.len = 0
        
    def push(self,val):
        if(self.Q1.full()):
            print("The Queue is full")
        else:
            self.Q1.put(val)
            self.len=self.len+1
            print("Value pushed into the stack : ",val)
        
    def pop(self):
        if(self.Q1.empty()):
            print("The Queue is empty")
        else:
            while(self.len != 1):
                self.Q2.put(self.Q1.get())
                self.len = self.len - 1
            
            self.temp = self.Q1
            self.Q1 = self.Q2
            self.Q2 = self.temp
            
            val = self.Q2.get()
            print("Value popped from the stack : ",val)

s1 = Stack_from_Queue()
s1.push(1)
s1.push(2)
s1.push(3)
s1.pop()
s1.push(22)
s1.pop()
```
